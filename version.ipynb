```python
import torch
from transformers import AutoProcessor, SiglipVisionModel
from PIL import Image

# 1. 模型名
model_name = "google/siglip-base-patch16-224"

# 2. 加载 processor 和 vision model
processor = AutoProcessor.from_pretrained(model_name)
model = SiglipVisionModel.from_pretrained(model_name)

# 3. 读一张图
image = Image.open("test.jpg").convert("RGB")

# 4. 预处理
inputs = processor(images=image, return_tensors="pt")

# inputs 里最重要的是 pixel_values
print("pixel_values shape:", inputs["pixel_values"].shape)   # [B, 3, H, W]

# 5. 前向
with torch.no_grad():
    outputs = model(pixel_values=inputs["pixel_values"])

# 6. 看输出
print("last_hidden_state shape:", outputs.last_hidden_state.shape)
print("pooler_output shape:", outputs.pooler_output.shape if outputs.pooler_output is not None else None)
```
